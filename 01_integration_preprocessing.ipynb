<h3>1. 데이터 파일 통합</h3>

```python
import pandas as pd

file_list = ['/content/drive/MyDrive/EDA project - 여행하소서/제출 파일/CI_OVSEA_TOUR_AREA_INTRST_DGREE_INFO_20230821.csv',
             '/content/drive/MyDrive/EDA project - 여행하소서/제출 파일/CI_OVSEA_TOUR_AREA_INTRST_DGREE_INFO_20230828.csv',
             '/content/drive/MyDrive/EDA project - 여행하소서/제출 파일/CI_OVSEA_TOUR_AREA_INTRST_DGREE_INFO_20230904.csv',
             '/content/drive/MyDrive/EDA project - 여행하소서/제출 파일/CI_OVSEA_TOUR_AREA_INTRST_DGREE_INFO_20230911.csv',
             '/content/drive/MyDrive/EDA project - 여행하소서/제출 파일/CI_OVSEA_TOUR_AREA_INTRST_DGREE_INFO_20230918.csv',
             '/content/drive/MyDrive/EDA project - 여행하소서/제출 파일/CI_OVSEA_TOUR_AREA_INTRST_DGREE_INFO_20230925.csv']

dfs = []
for file in file_list:
    df = pd.read_csv(file)
    dfs.append(df)

final_df = pd.concat(dfs, ignore_index=True)
final_df
```

<h3>2. 데이터 전처리</h3>
<h4>1) 영문 컬럼명 한글로 바꾸기</h4>

```python
cd = pd.read_excel('/content/drive/MyDrive/EDA project - 여행하소서/제출 파일/해외 여행지역 관심도_컬럼정의서.xls', index_col=0)
cd.head()
```

```python
column_mapping = cd.set_index('컬럼영문명')['컬럼한글명'].to_dict()
final_df.rename(columns=column_mapping, inplace=True)
final_df
```

<h4>2) 관심값 컬럼 레이블 인코딩</h4>

```python
def encode_intrst(x):
  if x == '많이 적어졌다':
    x = 1
  elif x == '약간 적어졌다':
    x = 2
  elif x == '예전과 비슷하다':
    x = 3
  elif x == '약간 커졌다':
    x = 4
  elif x == '많이 커졌다':
    x = 5
  return x

final_df[final_df.columns[-12:]] = final_df[final_df.columns[-12:]].applymap(encode_intrst)
final_df.head()
```

<h4>3) 소득수준 컬럼 레이블 인코딩</h4>

```python
import numpy as np
def encode_income(x):
  if x == '300만원 미만':
    x = 1
  elif x == '300이상500만원 미만':
    x = 2
  elif x == '500이상700만원 미만':
    x = 3
  elif x == '700만원 이상':
    x = 4
  elif x == '모름':
    x = np.nan
  return x

final_df['가구소득정도명'] = final_df['가구소득정도명'].apply(encode_income)
final_df.head()
```

<h4>4) 연령대 컬럼 integer로 변경</h4>

```python
final_df['연령대구분명'] = final_df['연령대구분명'].str.replace('대', '').astype(int)
final_df.head()
```

<h4>4) 결측치 채우기</h4>

```python
final_df[final_df['가구소득정도명'].isna()]
```

```python
# 연령대별 가구소득 최빈값으로 결측치 대체
income_by_age = final_df.groupby(['연령대구분명'])['가구소득정도명'].agg(lambda x: x.mode().iloc[0])
income_by_age
```

```python
final_df['가구소득정도명'] = final_df['가구소득정도명'].fillna(final_df['연령대구분명'].map(income_by_age))
final_df.head(50)
```
